```python
# get credentials from server date

response = requests.post(url)

from datetime import datetime
def generate_credentials(server_date_str):
    """Generate username dan password"""
    
    # Parse server date untuk DDMMYY
    server_date = datetime.strptime(server_date_str, '%a, %d %b %Y %H:%M:%S %Z')
    
    # Ambil waktu lokal untuk jam (C{jam})
    local_time = datetime.now()
    local_hour = local_time.strftime('%H')  # 2 digit jam lokal
    
    # Generate username: tesprogrammer{DDMMYY}C{jam_lokal}
    username = f"tesprogrammer{server_date.strftime('%d%m%y')}C{local_hour}"
    
    # Generate password: bisacoding-DD-MM-YY
    password = f"bisacoding-{server_date.strftime('%d-%m-%y')}"
    
    return username, password, server_date, local_time

server_date_str = response.headers['Date']
username, password, server_date, local_time = generate_credentials(server_date_str)

print(f"Server Date: {server_date_str}")
print(f"Waktu Server (untuk DDMMYY): {server_date}")
print(f"Waktu Lokal (untuk C{local_time.strftime('%H')}): {local_time}")
print(f"Username: {username}")
print(f"Password: {password}")
```

```python
import json
import hashlib
def get_data_api(username, password):
    params = {'username': username, 'password': hashlib.md5(password.encode()).hexdigest()}
    response = requests.post(url, data=params)
    return response.json()

data = get_data_api(username, password)
print(json.dumps(data, indent=4))
```
